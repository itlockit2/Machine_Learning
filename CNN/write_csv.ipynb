```python
import csv
import os
from numpy import array
from numpy import shape
```

```python
TRAIN_DIR = "/home/banana/Machine_Learning/CNN/MsCeleb/MsCelebV1-DevSet1/aligned"
TEST_DIR = "/home/banana/Machine_Learning/CNN/MsCeleb/MsCelebV1-DevSet2/aligned"
train_file_list = array(os.listdir(TRAIN_DIR))
test_file_list = array(os.listdir(TEST_DIR))
print(shape(train_file_list))
print(shape(test_file_list))
f = open('training_file.csv','w', encoding='utf-8', newline='')
wr = csv.writer(f)
count = 0
for i in range(len(train_file_list)):
    wr.writerow([TRAIN_DIR + '/'+ train_file_list[i],count])
    count = count + 1
    
f2 = open('test_file.csv','w', encoding='utf-8', newline='')
wr = csv.writer(f2)
count = 0
for i in range(len(test_file_list)):
    wr.writerow([TEST_DIR + '/' +test_file_list[i],count])
    print(TEST_DIR + '/' +test_file_list[i],count)
    count = count + 1
```
